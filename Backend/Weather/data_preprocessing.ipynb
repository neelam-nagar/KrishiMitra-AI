```python
import json

with open("final_clean.json", "r", encoding="utf-8") as f:
    data = json.load(f)

len(data["districts"])
```

```python
sample = data["districts"][0]

sample.keys(), sample["tehsils"][0].keys(), sample["tehsils"][0]["villages"][0].keys()
```

```python
missing = 0
total = 0

for d in data["districts"]:
    for t in d["tehsils"]:
        for v in t["villages"]:
            total += 1
            if not v.get("lat") or not v.get("lon"):
                missing += 1

total, missing
```

```python
from collections import Counter

names = []

for d in data["districts"]:
    for t in d["tehsils"]:
        for v in t["villages"]:
            names.append(v["village"].lower())

[x for x, c in Counter(names).items() if c > 1][:10]
```

```python
with open("final_ready.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("✅ FINAL DATASET READY")
```
